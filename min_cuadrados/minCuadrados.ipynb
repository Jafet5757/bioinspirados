```python
import pandas as pd
import matplotlib.pyplot as plt

# leemos el archivo de datos Salary.csv
data = pd.read_csv('Salary.csv')

data.describe()
```

El dataset elegido contiene datos de salarios y años de experiencia, estos tiene un comportamiento lineal y es más visible esto al graficarlos

```python
#Graficamos los datos
plt.scatter(data['YearsExperience'],data['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

Usar el metodo de minimos cuadrados nos ayudará a encontrar una recta que se ajuste a estos datos y pueda predecir su comportamiento

```python

# Obtenemos los valores de X y Y
x = data['YearsExperience']
y = data['Salary']

# Obtenemos la multiplicacion de la sumatoria de x * y
xy = x*y

# obtenemos x^2
x2 = x**2

# Obtenemos Sx*Sy
SxSy = x.sum()*y.sum()

# Obtenemos el total de datos
n = data['YearsExperience'].count()

# Calculamos la pendiente m
m = (xy.sum() - (SxSy/n))/(x2.sum() - ((x.sum()**2)/n))

# Calculamos la interseccion b
b = (y.sum()/n) - m*(x.sum()/n)

# Imprimimos los valores de m y b
print("m = ",m)
print("b = ",b)

# Graficamos la recta
plt.scatter(x,y)
plt.plot(x,m*x+b,color='red')
```
